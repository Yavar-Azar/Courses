```python
import numpy as np


def gaussian1(x, mu, sigma):
    
    gg=(1/(sigma*np.sqrt(2*np.pi)))*np.exp(-(x-mu)**2/(2*sigma**2))
    
    return gg
    
    
```

```python
x=np.linspace(-18, 22, 1000)

f= gaussian1(x, 2, 5)
```

```python
plt.plot(x, f)
```

```python
np.sum(f)*(x[1]-x[0])
```
