## Derivatives

5 point 


$$f'(x)\approx {\frac  {-f(x+2h)+8f(x+h)-8f(x-h)+f(x-2h)}{12h}}$$

or in other word:

$$f'(x)\approx {\frac  {-f_{+2}+8f_{+1}-8f_{-1}+f_{-2}}{12h}}$$


```python
import numpy as np 

x=np.linspace(-5,5, 100)

f1=np.sin(x)

#f2=x**2+4*x


def deriv5p(y,x):
    """
    y:array
    array of function 
    
    x:array
    range of x values
    """
    #  we have four terms based on definition then ..
    
    t2p = -y[4:]
    t1p = +8*y[3:-1]
    tz  = y[2:-2]
    t1m = -8*y[1:-3]
    t2m = +y[:-4]
    
    h=x[1]-x[0]
    result = (t2p+t1p+t1m+t2m)/(12*h)
    
    return result



    
    
    


```

```python
rs = deriv5p(f1,x)
```

```python
import matplotlib.pyplot as plt
dist = x[1]-x[0]

#  This is gradient by numpy lib
test = np.gradient(f1, dist)

plt.plot(x, f1)
plt.plot(x[2:-2], rs, lw=5, label='my function')
plt.plot(x,test, lw=1, label='numpy result', c='black')
```
